```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor()
model_10 = RandomForestRegressor(n_estimators = 10, random_state = 0)
model_10.fit(df_xtrain, df_ytrain)
```

Calculating the loss after training - Root Mean Square Error (RMSE)

```python
from sklearn.metrics import mean_squared_error
```

```python
pred_10 = model_10.predict(df_xtest)
rmse_10 = float(format(np.sqrt(mean_squared_error(df_ytest, pred_10)), '.3f'))
print("\nRMSE: ", rmse_10)
error_10 = df_ytest-pred_10
aerror_10 = np.abs(error_10)
MAPE_10=np.mean(aerror_10/df_ytest)*100
MAPE_10
accuracy_10 = 100 - np.mean(MAPE_10)
print('Accuracy:', round(accuracy_10, 2), '%.')
```
